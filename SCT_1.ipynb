```python
import pandas as pd

df = pd.read_csv("population_distribution.csv")
print(df)
```

```python
# Sum population by age group
age_distribution = df.groupby("Age Group")["Population"].sum().reset_index()
print(age_distribution)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.barplot(x="Age Group", y="Population", data=age_distribution, palette="muted")

plt.title("Population Distribution by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Total Population")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x="Population", bins=5, kde=True, color="skyblue")

plt.title("Histogram of Population Values")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
